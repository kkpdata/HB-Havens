```python
# Get resultdata for first added location in HRD
conn = sqlite3.connect(db2)
hrdlocationids1 = np.arange(734, 820)
hrdlocationids2 = np.arange(820, 906)
supportlocid = 567
hrdlocidsstr = ','.join(np.r_[supportlocid, hrdlocationids2].astype(str).tolist())


windspeedid = conn.execute('select HRDInputColumnId from HRDInputVariables where InputVariableId = 9').fetchall()[0][0]
waterlevelid = conn.execute('select HRDInputColumnId from HRDInputVariables where InputVariableId = 10').fetchall()[0][0]

SQLs = {
    'Wind speed': f"""
        select HRDLocationId, ID.Value, Count(HD.HydroDynamicDataId) from HydroDynamicData HD
        inner join HydroDynamicInputData ID on ID.HydroDynamicDataId = HD.HydroDynamicDataId
        where ID.HRDInputColumnId = {windspeedid} and HRDLocationId IN ({hrdlocidsstr})
        group by HRDLocationId, ID.Value""",
    'Water level': f"""
        select HRDLocationId, ID.Value, Count(HD.HydroDynamicDataId) from HydroDynamicData HD
        inner join HydroDynamicInputData ID on ID.HydroDynamicDataId = HD.HydroDynamicDataId
        where ID.HRDInputColumnId = {waterlevelid} and HRDLocationId IN ({hrdlocidsstr})
        group by HRDLocationId, ID.Value""",
    'Wind direction': f"""
        select HRDLocationId, Direction as Value, Count(HydroDynamicDataId) from HydroDynamicData
        inner join HRDWindDirections on HRDWindDirections.HRDWindDirectionID = HydroDynamicData.HRDWindDirectionID
        where HRDLocationId IN ({hrdlocidsstr})
        group by HRDLocationId, HydroDynamicData.HRDWindDirectionId"""
}

# Test the number of entries per input column
input_columns = ['Wind speed', 'Water level', 'Wind direction']

# Counts from db
for col in input_columns:
    counts = pd.read_sql(SQLs[col], con=conn, index_col=['HRDLocationId', 'Value']).unstack()

    check_count = counts.loc[supportlocid]
    faulty = counts.index[~(counts == check_count).all(axis=1).values].tolist()

    print(faulty)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, 100, 40)
y = np.linspace(0, 20, 10)
X, Y = np.meshgrid(x, y)
```

```python
trange = np.linspace(0, 2* np.pi, 20)
overal = 3

for t in trange:
    
    fig, ax = plt.subplots()
    ax.set_aspect(1.0)
    ax.set_ylim(-3, 30)
    
    dt = np.pi * x * 0.02 * overal
    
    hfactor = np.exp(-y[::-1] * 0.125)[:, None] * overal
    
    Xwave = X + np.cos(dt + t) * hfactor + dt
    Ywave = Y + np.sin(dt + t) * hfactor
    
    ax.plot(Xwave, Ywave, marker='.', ls='', color='C0', ms=2);
    
    ax.fill(np.r_[Xwave[0], Xwave[-1][::-1]], np.r_[Ywave[0], Ywave[-1][::-1]], alpha=0.1)

    
```
